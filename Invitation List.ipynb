```python
import requests
import json
from datetime import date, timedelta

res = requests.get('https://ct-api-challenge.herokuapp.com/')
data = res.json()

country_Dict = {}
country_List=  []

for i in range(len(data['partners'])):
    if data['partners'][i]['country'] not in country_List:
        country_Dict[data['partners'][i]['country']] = {}
        country_List.append(data['partners'][i]['country'])
        

counter = 0
used_List = []

for i in range(len(data['partners'])):
    first_Name       = data['partners'][i]['firstName']
    last_Name        = data['partners'][i]['lastName']
    email_Add        = data['partners'][i]['email']
    given_Dates      = data['partners'][i]['availableDates']
    if data['partners'][i]['country'] not in used_List:
        counter = 0
        used_List.append(data['partners'][i]['country'])
    counter+=1
    country_Dict[data['partners'][i]['country']][counter] = {'First Name'      : first_Name,
                                                             'Last Name'       : last_Name,
                                                             'Email Address'   : email_Add,
                                                             'Given Dates'     : given_Dates
                                                            }

```

```python
import datetime

for country in country_List:
    #print('\n', country)
    list1   = []
    for i in range(len(data['partners'])):
        for date in data['partners'][i]['availableDates']:
            if data['partners'][i]['country'] == country:
                list1.append(date)

    checked_Dates  = []
    date_Occurence1 = []
    date_Occurence2 = []
    
    max            = 0

    
    for date in list1:
        occurence = 0
        if date not in checked_Dates:
            checked_Dates.append(date)
            occurence = list1.count(date)

            if occurence >= max:
                date_Occurence1.append((date,occurence))
                max = occurence


          
    for i in date_Occurence1:
        d = timedelta(1)
        day = datetime.date(int(i[0][:4]), int(i[0][5:7]), int(i[0][8:10]))    
        difference = datetime.timedelta(days=1)
        day_after = day + difference 

        for j in date_Occurence1:   
            if str(day_after) == j[0]:
                date_Occurence2.append(i)

                
    country_Dict[country]['Available Dates'] = date_Occurence2            
```

```python
import datetime

for key,value in country_Dict.items():
    print('\n'+"*"*50)
    print('\nCountry:      ', key)
    print('\nStart Date:   ', value['Available Dates'][-1][0])
    print("\nAttendeeCount:", value['Available Dates'][-1][-1])
    print('\nAttendees:\n')

    day        = datetime.date(int(value['Available Dates'][-1][0][:4]), int(value['Available Dates'][-1][0][5:7]), int(value['Available Dates'][-1][0][8:10]))    
    difference = datetime.timedelta(days=1)
    day_after = day + difference 
    
    for i in range(1,len(value)):
        if str(day) in value[i]['Given Dates'] and str(day_after) in value[i]['Given Dates']:
            print("              ", value[i]['First Name'], value[i]['Last Name'])
     
```
